    MultiGridplots

FacetGrid

```python
g= sns.FacetGrid(data=tips,hue='smoker',col='day',row='time')
g.map(sns.violinplot,'sex','total_bill')
g.add_legend()
```

```python
g= sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.boxplot,'sex','total_bill')  
g.add_legend()
```

Pairplot

```python
iris.head()
```

```python
sns.pairplot(data=iris,hue='species')
plt.show()
```

Pairgrid

```python
g=sns.PairGrid(data=iris,hue='species')
g.map(sns.scatterplot)
```

```python
#diagoanl and offdiagonal
g=sns.PairGrid(data=iris,hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
#map_diag -> map_lower
g=sns.PairGrid(data=iris,hue='species')
g.map_lower(sns.scatterplot)
```

```python
#map_diag -> map_upper
g=sns.PairGrid(data=iris,hue='species')
g.map_upper(sns.scatterplot)
```

```python
#map_diag -> map_lower + map_upper
g=sns.PairGrid(data=iris,hue='species')
g.map_lower(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
```

```python
#vars
g=sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_lower(sns.scatterplot)
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
```
